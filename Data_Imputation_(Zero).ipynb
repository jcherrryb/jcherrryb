# **DATA IMPUTATION**

```python
import pandas as pd
```

```python
df1 = pd.read_csv('notime_weather.csv')
df2 = pd.read_csv('transport_ridership.csv')
merged_df = pd.merge(df1, df2, on='date', how='inner')
```

```python
#Drop irrelevant weather columns (adjust based on your data)
columns_to_drop = [
    'pressure', 'dew_point', 'gust',
    'wind_chill', 'uv_index', 'humidity', 'wind_speed',
    'visibility', 'pollutant_value', 'precipitation_rate', 'feels_like_temperature',
    'bus_rkn', 'bus_rpn', 'rail_ets', 'rail_intercity', 'rail_komuter_utara', 'rail_tebrau',
    'rail_komuter'
]
merged_df = merged_df.drop(columns=columns_to_drop, errors='ignore')
```

```python
#Rename Weather Columns for Simplicity ===
merged_df = merged_df.rename(columns={
    'temperature': 'hotness',
    'solar_radiation': 'sunlight',
    'precipitation_total': 'rain'
})
```

```python
#Print the missing values
print("\Missing Values:\n", missing_values)
```

```python
#Impute Missing Values
#Numerical columns
num_cols = ['hotness', 'sunlight', 'rain']
for col in num_cols:
    if col in merged_df.columns:
        merged_df[col] = merged_df[col].fillna(merged_df[col].median())

#Categorical columns
if 'station' in merged_df.columns:
    merged_df['station'] = merged_df['station'].fillna(merged_df['station'].mode()[0])
```

```python
#List of important columns
important_cols = [
    'place', 'city', 'state']

#List of transport ridership columns
ridership_cols = [
    'bus_rkl',
    'rail_lrt_ampang', 'rail_mrt_kajang', 'rail_lrt_kj',
    'rail_monorail', 'rail_mrt_pjy'
]

#Fill missing ridership values with 0 (the transport didn't start the operation yet)
for col in ridership_cols:
    if col in merged_df.columns:
        merged_df[col] = merged_df[col].fillna(0).astype(int)
```

```python
#Final Check
print("Missing values (after cleaning + imputing):")
print(merged_df.isnull().sum())
```
